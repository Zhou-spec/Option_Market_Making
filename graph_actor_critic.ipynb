```python
import matplotlib.pyplot as plt
%matplotlib ipympl
import numpy as np
```

```python
loss_v = np.load('loss_v.npy')
loss_p = np.load('loss_p.npy')

x = [loss_v[i] / loss_v[0] for i in range(len(loss_v))]
y = [loss_p[i] / loss_p[0] for i in range(len(loss_p))]
# Plot the loss
plt.figure()
plt.plot(x, label='loss_v')
plt.legend()
plt.savefig('loss_v_ac.png')
plt.show()

plt.figure()
plt.plot(y, label='loss_p')
plt.legend()
plt.savefig('loss_p_ac.png')
plt.show()

```

```python
from scipy.stats import gaussian_kde
f = np.load('final_returns_ac.npy')

kde = gaussian_kde(f)
x = np.linspace(-100, 200, 100)
plt.figure(figsize=(10, 5))
plt.hist(f, bins = 20, edgecolor='black', density=True)
plt.plot(x, kde(x), color='brown', alpha = 0.7)
plt.savefig('final_returns_ac.png')
plt.show()
```
